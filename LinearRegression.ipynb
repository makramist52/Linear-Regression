```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.linear_model import LinearRegression
```

```python
# Sample data
X = np.array([[1], [2], [3], [4], [5]])  # Input features
y = np.array([3, 4, 2, 4, 5])           # Target variable

# Create a linear regression model
# Creates an instance of the LinearRegression model.
model = LinearRegression()

# Fit the model to the data
#Trains the linear regression model using the input features X and target variable y.
model.fit(X, y)

# Predict using the trained model

X_test = np.array([[6], [7], [8]])  # New input features
y_pred = model.predict(X_test)

# Predicted target variable.Prints the predicted target variable values.

# Print the coefficients and intercept
#In this specific line, model.coef_[0] retrieves the coefficient from the trained linear regression model (model). 
#The [0] index is used because coef_ is an array, and in this case, we have only one input feature. 
# Therefore, it prints the coefficient value associated with the single input feature.
#The printed output will display the coefficient value, providing insight into the relationship 
#between the input feature and the target variable in the linear regression model.

print("Coefficient:", model.coef_[0])

#the intercept represents a constant term in the regression equation and is typically a numeric value
# retrieves the intercept value from the trained linear regression model (model).

#The printed output will display the intercept value, providing insight into the baseline value of the 
#target variable in the linear regression model, regardless of the input feature values.

print("Intercept:", model.intercept_)

# Print the predicted values
#the code is printing the predicted values of the target variable based on the trained linear regression model.

print("Predicted values:", y_pred)

# Plot the data points
plt.scatter(X, y, color='blue', label='Data')

# Plot the linear regression line
plt.plot(X, model.predict(X), color='red', label='Linear Regression')

# Add labels and title to the plot
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression')

# Add a legend
#A legend is a key that provides information about the elements in the plot, 
#such as the data points and lines.
#It helps in identifying different components of the plot and understanding their meaning.

plt.legend()

# Display the plot
plt.show()
```
